```python
#WITH RE
import re
import time
start_time = time.time()


f = open("emails.txt", "r")

emails = f.read()
#[a-zA-Z0-9_.-]+[^!#$%^&*()]+@gmail\.com|@yahoo\.com|@hotmail\.com|@outlook\.com

#[\w]+@gmail\.com|[\w]+@hotmail\.com|[\w]+@outlook\.com|[\w]+@yahoo\.com


email_re = r'[a-zA-Z0-9_.-]+[^!#$%^&*()]@yahoo+\.\bcom|[a-zA-Z0-9_.-]+[^!#$%^&*()]@gmail+\.\bcom|[a-zA-Z0-9_.-]+[^!#$%^&*()]@hotmail+\.\bcom|[a-zA-Z0-9_.-]+[^!#$%^&*()]@outlook+\.\bcom'
pattern = re.compile(email_re)

matches = pattern.findall(emails)
x=0;
for match in matches:
    x=x+1
    print(match)
    


print(x)
print("--- %s seconds ---" % (time.time() - start_time))
```
